```python
from sage.all import *
import pickle
import time
from sage.all import sigma, nth_prime, matrix, QQ, polygen, log, list_plot, flatten, pari
start=time.time()
precision_bits = 100
CC = ComplexField(precision_bits)
x = polygen(QQ)

def h_sequence_from_j(j_list):
    j_lenth = len(j_list)
    h = [QQ(1)]   # h_0 = 1
    for n in range(1, j_lenth + 1):
        s = sum(j_list[r-1] * h[n-r] for r in range(1, n))
        h_n = (j_list[n-1] + s) / n
        h.append(h_n)
    return h

def j_sequence_from_h(h_list):
    if h_list[0] != 1:
        raise ValueError("Must have h_0 = 1")
    h_lenth= len(h_list) - 1
    j = []
    for n in range(1, h_lenth + 1):
        s = sum(j[r-1] * h_list[n-r]
                for r in range(1, n))
        j_n = n*h_list[n] - s
        j.append(j_n)
    return j

h_list=[1]+[2 for k in range(1,101)]
j_list=j_sequence_from_h(h_list)
deformed_j_list=[1]+j_list
deformed_h_list= h_sequence_from_j(deformed_j_list)
print(h_list[:20])
print()
print(j_list[:20])
print()
print("deformed_j_list:")
print(deformed_j_list[:20])
print()
print("deformed_h_list:")
print(deformed_h_list[:20])
print()
nums=[numerator(x) for x in deformed_h_list]
print("numerators deformed_h_list:")
print(nums[:20])
# odd n: A354334 a(n) is the numerator of Sum_{k=0..n} 1 / (2*k)!
# = Numerators of coefficients in expansion of cosh(sqrt(x)) / (1 - x).
# even n: A354211 a(n) is the numerator of Sum_{k=0..n} 1 / (2*k+1)!
# = Numerators of coefficients in expansion of sinh(sqrt(x)) / (sqrt(x) * (1 - x)).
print()
dens=[denominator(x) for x in deformed_h_list]
print("denominators deformed_h_list:")
print(dens[:20])
# even n: A354331 a(n) is the denominator of Sum_{k=0..n} 1 / (2*k+1)!
# = Denominators of coefficients in expansion of sinh(sqrt(x)) / (sqrt(x) * (1 - x)).
# EXAMPLE 1, 7/6, 47/40, 5923/5040, 426457/362880, 15636757/13305600, 7318002277/6227020800, ...
# (these are the odd entries in the deformed h_list)
# odd n: A354335 a(n) is the denominator of Sum_{k=0..n} 1 / (2*k)!m
# = Denominators of coefficients in expansion of sinh(sqrt(x)) / (sqrt(x) * (1 - x)).
```
