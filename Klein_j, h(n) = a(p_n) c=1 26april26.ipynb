```python
import pickle
from sage.all import *
import time

start = time.time()

R.<q> = PowerSeriesRing(QQ)
j = j_invariant_qexp(prec=2900)
klein=j.list()
print("klein1:")
print(klein[:10])

klein_list =[] 
for k in range(-1, 2800):
    klein_list.append([k-1,klein[k]])
    #if k%50==0:
       # print("klein",k)

print("short klein_list:")
print(klein_list[1:10])

with open('/Users/barrybrent/data2/run26april26no1.txt', 'wb') as wfile:
    pickle.dump(klein_list, wfile) #output snipped

klein_on_primes=[pair for pair in klein_list if is_prime(pair[0])]
short_primes=klein_on_primes[:10]
print()
print("short primes:")
print(short_primes)

h_list=[(0,1)]+klein_on_primes 
h_vals=[pair[1] for pair in h_list]
j_list=[(0,1)] # <<< c=1
for n in range(1,401):
    sm=0
    for r in [1..n-1]: 
        sm=sm+j_list[r][1]*h_vals[n-r]
    j_n=n*h_list[n][1]-sm
    j_list.append((n,j_n))
    if n%100==0:
        print("j_list:",n)

    
with open('/Users/barrybrent/data2/run26april26no2.txt', 'wb') as wfile:
    pickle.dump(j_list, wfile) # ouput snipped

now=time.time()
print("elapsed seconds:",now-start)
```

```python
import pickle
from sage.all import sigma, nth_prime
import ast
import time

start=time.time()
with open('/Users/barrybrent/data2/run26april26no2.txt', 'rb') as rfile:
    j_list = pickle.load(rfile)
    
def P_matrix(lst):
    lenlist=len(lst)
    columns=[lst]
    for k in range(1,lenlist):
        column=[0]*(k-1)
        column=column+[-k]
        column=flatten(column+[lst])
        column=column+[lst]
        column=column[:lenlist]
        columns+=[column]
    return (matrix(columns)).transpose()

from sage.all import *
precision_bits = 100
CC = ComplexField(precision_bits)
x = polygen(QQ)  # Can remain in preamble if all polynomials are in one variable

charpol_list=[]
roots_list=[]

mins_list=[]
log_mins_list=[]

dets=[]

for n in range(2, 201):
    now=time.time()
    print(n,(now-start)/60)
    deformed_j_list=j_list[:n]#<<<<<<<<<<<<<<<<<<<<<<<< deformed
    inputs=[pair[1] for pair in deformed_j_list]
    pl=P_matrix(inputs) 
    det=pl.det()
    dets.append((n,det))
    cp=pl.charpoly()
    charpol_list.append((n,cp))
    roots = cp.roots(ring=CC, multiplicities=False)       
    roots_list.append((n,roots))
    aar = [r.abs() for r in roots]
    if aar!=[]:
        minn = min(aar)  
        mins_list.append((n,minn))
        
        

    if n%20==0:
        p = list_plot(mins_list, plotjoined=True, axes_labels=['', ''])
        print("MINIMUM MODULI")
        p.show()

p = list_plot(mins_list, plotjoined=True, axes_labels=['', ''])
print("MINIMUM MODULI")
p.show()

with open('/Users/barrybrent/data2/run26april26no3.txt', 'wb') as wfile:
    pickle.dump(roots_list, wfile) #output snipped

with open('/Users/barrybrent/data2/run26april26no4.txt', 'wb') as wfile:
    pickle.dump(charpol_list, wfile) #output snipped

with open('/Users/barrybrent/data2/run26april26no5.txt', 'wb') as wfile:
    pickle.dump(mins_list, wfile) #output snipped
```
