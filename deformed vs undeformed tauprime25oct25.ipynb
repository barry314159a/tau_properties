```python
for offset in [2..50]:
    lst=[]
    shifts=[]
    print("offset:",offset)
    
    for k in range(0,200):
        deformed_poly=deformed_charpol_list[k][1]
        cff=deformed_poly[k-offset]
        lst+=[cff]

    ratios=[]
    for k in [offset..200-offset]:
        ratios+=[lst[k+1]/lst[k]]
    nums=[numerator(a) for a in ratios]
    dens=[denominator(a) for a in ratios]
    shifts=[nums[a]-dens[a+offset+1] for a in [0..-(offset+2)+min(len(nums),len(dens))]]
    print("lst[",offset,"]:",lst[offset])
    print()
    print("tests:")
    def gcd_product(n):
        return gcd(n,1+offset)
    tests=[n/gcd_product(n) for n in [1..198]]
    mn=min(len(tests),len(dens))
    new_tests=[tests[n]-dens[n] for n in range(0,mn)]
    print(new_tests)  
    print()
    print("shifts:")
    print(shifts)
    print("=================================================================================================")
```
