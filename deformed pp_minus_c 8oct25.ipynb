```python
import pickle
from sage.all import sigma, nth_prime
import ast
import math

    
def P_matrix(lst):
    lenlist=len(lst)
    columns=[lst]
    for k in range(1,lenlist):
        column=[0]*(k-1)
        column=column+[-k]
        column=flatten(column+[lst])
        column=column+[lst]
        column=column[:lenlist]
        columns+=[column]
    return (matrix(columns)).transpose()

from sage.all import *
precision_bits = 100
CIF = ComplexIntervalField(precision_bits)
x = polygen(QQ)  # Can remain in preamble if all polynomials are in one variable
main_matrix_list=[]
main_charpol_list=[]
main_roots_list=[]
main_mins_list=[]
main_maxes_list=[]
main_dets=[]
main_j_list=[]
for c in range (-10, 11):
    matrix_list=[]
    charpol_list=[]
    roots_list=[]
    aar_list=[]
    centers_list=[]
    max_diameter_list=[]
    mins_list=[]
    log_mins_list=[]
    maxes_list=[]
    log_maxes_list=[]
    dets=[]
    tests=[]
    def pph(k):
        return nth_prime(k+1)-nth_prime(k)-c
    pph_list=[]
    for k in range(1, 501):
        pph_list.append([k,pph(k)])
    
    h_list=[[0,1]]+pph_list # h_0 = 1 required by the lemmas
    j_list=[(0,0)] # first entry is just an offset. 
    for k in [1..500]:# so j(1) = 1 = h(0) as required.
        sm=0
        for r in [1..k-1]: 
            sm=sm+j_list[r][1]*h_list[k-r][1] 
        j_k=k*h_list[k][1]-sm
        j_list.append((k,j_k))
    main_j_list.append((c,j_list))
    for n in range(1, 201):
        deformed_j_list=j_list[:n]
        inputs=[pair[1] for pair in deformed_j_list]
        pl=P_matrix(inputs)
        det=pl.det()
        print((c,n))
        dets.append((n,det))
        matrix_list.append((n,pl))
        cp=pl.charpoly()
        charpol_list.append((n,cp))
        roots = cp.roots(ring=CIF, multiplicities=False)       
        roots_list.append((n,roots))
        centers=[r.center() for r in roots]
        pairs = [(z.real(), z.imag()) for z in roots]  
        aar = [r.abs() for r in roots]
        aar_list.append((n,aar))
    # compute absolute value for each root (before any conversion)
        if aar!=[]:
            minn = min(aar)  
            mins_list.append((n,minn))
            maxx = max(aar)  
            maxes_list.append((n,maxx))
            
            
    main_matrix_list.append((c,matrix_list))
    main_charpol_list.append((c,charpol_list))
    main_roots_list.append((c,roots_list))
    main_mins_list.append((c,mins_list))
    main_maxes_list.append((c,maxes_list))
   
#print("tests:")
#print(tests)

    # Filter out NaN and Inf before plotting
    maxes_list_clean = [(n, val) for n, val in maxes_list 
                        if math.isfinite(float(val))]
    log_maxes_list_clean = [(n, val) for n, val in log_maxes_list 
                            if math.isfinite(float(val))]
    

    
    p = list_plot(mins_list, plotjoined=True, axes_labels=['', ''])
    print("MINIMUM MODULI")
    p.show()

    p = list_plot(maxes_list, plotjoined=True, axes_labels=['', ''])
    print("MAXIMUM MODULI")
    p.show()
    
        
with open('/Users/barrybrent/data/run8oct25no19.txt', 'wb') as wfile:
        pickle.dump(main_roots_list, wfile) #output snipped

with open('/Users/barrybrent/data/run8oct25no20.txt', 'wb') as wfile:
    pickle.dump(main_charpol_list, wfile) #output snipped

with open('/Users/barrybrent/data/run8oct25no21.txt', 'wb') as wfile:
    pickle.dump(main_mins_list, wfile) #output snipped

with open('/Users/barrybrent/data/run8oct25no22.txt', 'wb') as wfile:
    pickle.dump(main_dets, wfile) #output snipped
    
with open('/Users/barrybrent/data/run8oct25no23.txt', 'wb') as wfile:
    pickle.dump(main_maxes_list, wfile) #output snipped

with open('/Users/barrybrent/data/run8oct25no24.txt', 'wb') as wfile:
    pickle.dump(main_j_list, wfile) #output snipped
```
